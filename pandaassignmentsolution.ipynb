```python
## adding in_stock column to inventory

inventory = pd.DataFrame(inventory, columns = ['location', 'product_type', 'product_description', 'quantity', 'price', 'in_stock'])


inventory['in_stock'] = np.where(inventory['quantity']!= 0, True, False)

display(inventory)
    
   
```

```python
# Adding new column total_value

inventory = pd.DataFrame(inventory, columns = ['location', 'product_type', 'product_description', 'quantity', 'price', 'in_stock', 'total_value'])

inventory['total_value']= inventory['price'] * inventory['quantity']

inventory
```

```python
## creating full description column

combine_lambda = lambda row: \
    '{} - {}'.format(row.product_type,
                     row.product_description)

inventory = pd.DataFrame(inventory, columns = ['location', 'product_type', 'product_description', 'quantity', 'price', 'in_stock', 'total_value', 'full_description'])

for i in inventory.index:
    inventory.loc[i, 'full_description'] = combine_lambda(inventory.loc[i, :])


display(inventory)
```
